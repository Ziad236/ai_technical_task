```python
import torch
from PIL import Image
import gradio as gr
from transformers import AutoTokenizer, AutoProcessor, AutoModelForCausalLM, GenerationConfig

model_id = "Qwen/Qwen-VL-Chat-Int4"

print("🚀 Loading model...")
tokenizer = AutoTokenizer.from_pretrained(model_id, trust_remote_code=True)
processor = AutoProcessor.from_pretrained(model_id, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="cuda",  # or "auto"
    trust_remote_code=True
).eval()
model.generation_config = GenerationConfig.from_pretrained(model_id, trust_remote_code=True)
print("✅ Model loaded!")
```

```python
def describe_image(image_path, temperature=0.5, top_p=0.9, max_new_tokens=300):
    try:
        image_url = image_path if image_path.startswith("http") else f"<img>{image_path}</img>"
        query = f"{image_url}Describe the image in detail."
        response, _ = model.chat(
            tokenizer,
            query=query,
            history=None,
            temperature=temperature,  # adjust creativity
            top_p=top_p,             # nucleus sampling
            max_new_tokens=max_new_tokens
        )
        return response
    except Exception as e:
        return f"❌ Error: {str(e)}"

# Gradio interface
with gr.Blocks() as demo:
    gr.Markdown("## 🧠 Qwen-VL-Chat-Int4 (AutoGPTQ) Image Captioning")

    with gr.Row():
        img_input = gr.Image(type="filepath", label="Upload Image")

    with gr.Row():
        with gr.Column():
            with gr.Accordion("Generation Parameters", open=False):
                temperature = gr.Slider(
                    minimum=0.1,
                    maximum=1.0,
                    value=0.5,
                    step=0.1,
                    label="Temperature (higher = more creative)"
                )
                top_p = gr.Slider(
                    minimum=0.1,
                    maximum=1.0,
                    value=0.9,
                    step=0.05,
                    label="Top-p (nucleus sampling)"
                )
                max_new_tokens = gr.Slider(
                    minimum=100,
                    maximum=500,
                    value=300,
                    step=50,
                    label="Max Description Length (tokens)"
                )
            describe_btn = gr.Button("Describe Image", variant="primary")

        with gr.Column():
            output = gr.Textbox(lines=8, label="Image Description")

    describe_btn.click(
        describe_image,
        inputs=[img_input, temperature, top_p, max_new_tokens],
        outputs=output
    )

demo.launch(share=True, debug=True)
```
